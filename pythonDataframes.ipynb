```python
#passing a dictionary below- where values are lists for each key
frame1=pd.DataFrame({'state':['Texas','Texas','Texas','Ohio','Ohio','Ohio'],
                    'population':[12000,12001,12002,50000,50001,50002],
                    'year':[2008,2009,2010,2008,2009,2010]})
frame1 #note, dictionary keys will become column heading, rows are associated with default index 
```

```python
#if you want to change the indices of a dataframe
frame1.index = ['a','b','c','d','e','f']
frame1
```

```python
#if you want to arrange columns,customize indices at the same time while creating the dataframe
frame1=pd.DataFrame({'state':['Texas','Texas','Texas','Ohio','Ohio','Ohio'],
                    'population':[12000,12001,12002,50000,50001,50002],
                    'year':[2008,2009,2010,2008,2009,2010]}, columns=['year', 'state', 'population'], index=['a','b','c','d','e','f'])
frame1
```

```python
#print first 5 rows of the dataframe
frame1.head()
```

```python
#print last 5 rows of the dataframe
frame1.tail() 
```

```python
#head() and tail() methods print 5 rows by default. If you want to print a specific number of rows, you can mention that in head() or tail() method
frame1.head(3)
```

```python
#check if values of a column are greater than a certain value
frame1['population']>30000
```

```python
#dataframe slicing-print one column
frame1['year']
```

```python
#print specific columns
frame1[['state','year']]
```

```python
#print only the first two rows (you need to use the original indices)
frame1[0:2]
```

```python
#slicing one row and two specific columns-you need to use loc function
frame1.loc['a',['state','year']]
```

```python
#slice two specific columns - output is same as printing two specific columns
frame1.loc[:,['state','year']]
```

```python
#now we are creating the dataframe with same values as above with arranged columns,customized indices
#where one column name is different. Note, I changed 'population' to 'pop' in arranged columns
frame2=pd.DataFrame({'state':['Texas','Texas','Texas','Ohio','Ohio','Ohio'],
                    'population':[12000,12001,12002,50000,50001,50002],
                    'year':[2008,2009,2010,2008,2009,2010]}, columns=['year', 'state', 'pop'], index=['a','b','c','d','e','f'])
frame2 # in output you see 'NAN' i.e. null/missing values for 'pop' column because the code is not able to recognize values for column heading 'pop' as 'pop' does not exist in the dictionary we used to create this dataframe 
```

```python
#check null values in a dataframe
pd.isnull(frame2) #see 'pop' column has all 'True' because this column has all null values
```

```python
#check not null values in a dataframe
pd.notnull(frame2)#see 'year' and 'state' columns have all 'True' because these columns have not null values
```

```python
#check summary statistics of data of a dataframe (this gives summary statistics of numerical columns of a dataframe)
#std stands for stanard deviation
frame1.describe()
```

```python
#if you want to check mean/average of a particular column
frame1['population'].mean()
```

```python
#add a new column to a dataframe
#we are adding a new column 'population_square' where the values will be the square of values of the 'population' column
frame1['population_square']= frame1['population']**2
frame1
```

```python
#delete a column from a dataframe
del frame1['population_square']
frame1
```

```python
#add a new row to a dataframe
frame1.loc['g']=[2012,'Ohio',12000]#write values of the new row as a list. maintain the sequence as per the column headings
frame1
```

```python
#delete a row
frame1.drop('g')
```

```python
#delete specific rows
frame1.drop(['c','d'])
```

```python
#overwrite the values of a row. note, we have overwritten the values for row 'a'
frame1.loc['a']=[2012,'Ohio',12000]
frame1
```

```python
#add a new column to a specific position
frame1.insert(2, "pop_square", frame1['population']**2, True)
frame1
```

```python
#sort dataframe by a specific column in ascending order
frame1=frame1.sort_values(by=['population'],ascending=True)
frame1#answer is sorted in ascending order by 'Salary' column values
```

```python
#b
data.head(3)
```

```python
#c
data.loc[['a','c','e'], ['Employee','Salary']]
```

```python
#2
data.insert(2, "Status", ['yes','yes','yes','yes','no','no'], True)
data
```

```python
data['Status']=data['Status'].map({'yes': True, 'no': False})
data
```

```python
#e
data.sort_values(by=['Salary'],ascending=True)
```

```python
data.sort_values(by=['Dept','Salary'],ascending=[False,False])
```

```python
#remove duplicate records
data_clean=data.drop_duplicates()
data_clean
```

```python
#g
data2= data_clean.loc[:,['Dept','Salary','Interest']]
data2
```
